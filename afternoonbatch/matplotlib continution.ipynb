```python
import matplotlib.pyplot as plt
```

```python
x  = np.arange(10)
y  = x**2
plt.plot(x,y)
plt.show()
```

```python
x  = np.arange(10)
y  = x**2
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title(" Y = X**2 ")
plt.plot(x,y)
plt.show()
```

```python
x  = np.arange(10)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title(" Line Plot ")
plt.plot(x,x,label = "Line Plot")
plt.plot(x,x**2,label = "Square Plot")
plt.plot(x,x**3,label = "Cube plot")
plt.legend()
plt.show()
```

```python
x  = np.arange(10)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title(" Line Plot ")
plt.plot(x,x,label = "Line Plot",linestyle = "--")
plt.plot(x,x**2,label = "Square Plot",linestyle= "-.")
plt.plot(x,x**3,label = "Cube plot",linestyle=":")
plt.legend()
plt.show()
```

## Bar chart

```python
x  = np.arange(1,100,2)
y  = np.linspace(1,50,len(x))
print(max(x),max(y))
print(min(x),min(y))
plt.xlabel("distance in m")
plt.ylabel("speed in km/hr")
plt.xlim(40,80)
plt.ylim(20,70)
plt.bar(x,y)
plt.show()
```

```python
#  bar chat
x=[20,36,58,49,59,30]
y=[5000,3990,4000,399,23890,6000]
#plt.tight_layout()
plt.bar(x,y,label='y-axis',color='y')
plt.title("basic bar gragh")
plt.xlabel("ages")
plt.ylabel("salary")
plt.grid(True)
plt.legend()
print(plt.style.available)
plt.show()

```

## Pie chart

```python
x  = [90,98,78,67,56,78]
plt.pie(x,labels = ["Python","c","c++","java","R","Android"],colors=["g","b","m","red","c","gray"])
plt.legend()
plt.xlim(5,5)
#plt.xlim(5,4)  # distance print pie chart
plt.show()
```

```python
# pie chart:
import matplotlib.pyplot as plt
labels=['python','html','java','c','sql']
slices=[60000,29000,30000,40000,41000]
explode=[0,0,0.1,0,0]
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'black'},startangle=90,explode=explode,shadow=True)
plt.tight_layout()
plt.show()
```

```python
# pie chart:
import matplotlib.pyplot as plt
labels=['python','html','java','c','sql']
slices=[60000,29000,30000,40000,41000]
explode=[0,1,0,1,0]
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':"yellow"},startangle=80,explode=explode,shadow=True)
plt.tight_layout()
plt.show()
```

```python
# subplot
import numpy as np

x=np.linspace(1,10,100)
plt.figure(1)
plt.subplot(2,2,1)
plt.title("linear")
plt.plot(x,x)

plt.figure(1)

plt.subplot(2,2,2)
plt.title("cubic")
plt.plot(x,x*2)


plt.figure(2)
plt.subplot(2,2,1)
plt.title("log")
plt.plot(x,np.log(x))

plt.figure(3)
plt.subplot(2,2,2)
plt.title("Exponential")
plt.plot(x,x**2)
plt.show()







```
